```python
#funcion objetivo
import math

def fit(lista_cantidades,billetes):
    suma = 0
    cantidad_billetes = 0
    for posc in range(len(lista_cantidades)):
        suma += lista_cantidades[posc] * billetes[posc] 
        cantidad_billetes += lista_cantidades[posc]
    resultado = (abs(s-suma))
    if(resultado==0):
        #9*9 es porque es la cantidad maxima de billetes que pueden haber
        #en unas celdas de mas abajo setee que se pueda tener de 0 a 9 billetes por cada uno de los 9 valores nominales.
        #es decir, 9*9
        resultado = -9*9 + cantidad_billetes  
    return resultado
fit([0,0,0,0,0,0,5,0,0],[1,5,10,20,50,100,200,500,1000])
```
